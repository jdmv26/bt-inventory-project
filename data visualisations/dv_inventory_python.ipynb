```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# using the url method of having the csv file read just to familiarise myself
sheet_url = "https://docs.google.com/spreadsheets/d/1rlPp2wUFP4Ll8be7shZFfM3eYh8qGp9RsiXEk4roBDo/edit#gid=0"
url_1 = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')
```

```python
inventorydf = pd.read_csv(url_1)
```

```python
# file successfully read
inventorydf
```

```python
inventorydf.head()
```

```python
# I wanted make a bar chart using Python to practice
plt.bar("product_name", "product_price", data = inventorydf, color = "teal", edgecolor = "blue")
plt.xlabel("Product Name")
plt.ylabel("Price of Product ()")
plt.title("The price of each individual product")
plt.xticks(rotation = 90)
plt.show()
```

```python
# libraries imported for the histogram
import numpy as np
import seaborn as sns
```

```python
# wbdf is where bought df.  I assigned the where_bought column of the inventory dataframe to wbdf
wbdf = inventorydf["where_bought"]
```

```python
# A pie chart isn't as interactive here as it is on Tableau. A histogram was made instead.
sns.displot(wbdf, discrete=True, color="teal", edgecolor = "blue", aspect= 2, height= 5, kind= "hist").set(title = "The number of songs released per year")
plt.xticks(rotation=90)
```
